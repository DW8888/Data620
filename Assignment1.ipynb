## 1. Import Required Libraries

```python
# Import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx  as nx
from networkx.drawing.nx_pydot import graphviz_layout
```

## 2.Initialize the graph

```python
G = nx.Graph()
```

## 3. Node creation (nodes are nouns in our case characters)

```python
names = [
    "Marcus", "Lucius", "Gaius", "Tiberius", "Aurelius",
    "Octavia", "Claudia", "Flavia", "Cornelia", "Valeria"
]
```

## 4. Feeling edgy? lets define our edges

```python
# 3. Add edges (same structure as the original Krackhardt kite graph)
edges = [
    ("Marcus", "Lucius"), ("Marcus", "Gaius"),
    ("Lucius", "Tiberius"), ("Lucius", "Aurelius"),
    ("Gaius", "Tiberius"), ("Octavia", "Claudia"),
    ("Tiberius", "Octavia"),
     ("Aurelius", "Claudia"),("Claudia","Lucius"),
    ("Octavia", "Cornelia"),("Gaius", "Claudia"),
    ("Claudia", "Cornelia"),("Gaius", "Octavia"),("Gaius", "Cornelia"),
    ("Flavia", "Valeria"), ("Aurelius", "Flavia"),("Gaius", "Lucius"),("Tiberius", "Marcus")
]
G.add_edges_from(edges)
```

## 5. Visualizing our graph.

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightblue",
        node_size=600, font_size=10, font_weight="bold")
plt.title("Krackhardt Kite Graph (Latin Names)", fontsize=14)
plt.show()

```

## Lets print the graph dimensions

```python
# 5. Print basic info
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Nodes:", list(G.nodes()))
print("Edges:", list(G.edges()))
```
